```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_palette("tab10")
```

```python
# Load  data
df = pd.read_csv("shopping_behavior_updated[2].csv")

print("Quick peek at the data:")
display(df.head())

print("\nShape of the data (rows, columns):", df.shape)
print("\nColumn info:")
print(df.info())
```

```python
print("\nMissing values per column:")
missing = df.isna().sum().sort_values(ascending=False)
print(missing)

print("\nDuplicate rows:", df.duplicated().sum())

print("\nBasic stats for numeric columns:")
display(df.describe().T)

print("\nUnique values for key categorical fields:")
for col in ["Gender", "Category", "Location", "Season",
            "Subscription Status", "Shipping Type", "Payment Method"]:
    if col in df.columns:
        print(f"\n{col}: {df[col].nunique()} unique values")
        print(df[col].value_counts().head())
```

```python
print("\nAverage customer age and spending:")
if "Age" in df.columns and "Purchase Amount (USD)" in df.columns:
    avg_age = df["Age"].mean()
    avg_spend = df["Purchase Amount (USD)"].mean()
    print(f"- Typical customer age is about {avg_age:.1f} years.")
    print(f"- On average, a single purchase brings in around ${avg_spend:,.2f}.")

if "Gender" in df.columns:
    print("\nPurchase count by gender:")
    print(df["Gender"].value_counts())

if set(["Gender", "Purchase Amount (USD)"]).issubset(df.columns):
    print("\nAverage spend by gender:")
    print(df.groupby("Gender")["Purchase Amount (USD)"].mean().round(2))

if "Location" in df.columns:
    print("\nTop 5 locations by number of orders:")
    print(df["Location"].value_counts().head())
```

```python
if "Category" in df.columns:
    print("\nTop 10 bestselling categories (by order count):")
    print(df["Category"].value_counts().head(10))

if set(["Category", "Purchase Amount (USD)"]).issubset(df.columns):
    print("\nAverage order value by category:")
    print(df.groupby("Category")["Purchase Amount (USD)"]
            .mean()
            .sort_values(ascending=False)
            .round(2)
            .head(10))

if set(["Season", "Purchase Amount (USD)"]).issubset(df.columns):
    print("\nAverage order value by season:")
    print(df.groupby("Season")["Purchase Amount (USD)"]
            .mean()
            .round(2))
```

```python
if "Discount Applied" in df.columns:
    print("\nShare of orders with a discount applied:")
    discount_rate = (df["Discount Applied"] == "Yes").mean() * 100
    print(f"- About {discount_rate:.1f}% of orders use some discount.")

if set(["Discount Applied", "Purchase Amount (USD)"]).issubset(df.columns):
    print("\nAverage spend for discounted vs nondiscounted orders:")
    print(df.groupby("Discount Applied")["Purchase Amount (USD)"]
            .mean()
            .round(2))

if "Promo Code Used" in df.columns:
    print("\nTop promo codes by usage:")
    print(df["Promo Code Used"].value_counts().head())
```

```python
if "Previous Purchases" in df.columns:
    print("\nBasic stats for previous purchases:")
    print(df["Previous Purchases"].describe())

if set(["Previous Purchases", "Purchase Amount (USD)"]).issubset(df.columns):
    print("\nDo customers who bought more in the past also spend more now?")
    corr_prev_spend = df["Previous Purchases"].corr(df["Purchase Amount (USD)"])
    print(f"- Correlation between past purchases and current spend: {corr_prev_spend:.3f}")
    if corr_prev_spend > 0:
        print("  This suggests loyal customers tend to place slightly higher orders.")
```

```python
if "Review Rating" in df.columns:
    print("\nReview score summary:")
    print(df["Review Rating"].describe())

if set(["Review Rating", "Purchase Amount (USD)"]).issubset(df.columns):
    print("\nDo highervalue orders get better ratings?")
    corr_rating_spend = df["Review Rating"].corr(df["Purchase Amount (USD)"])
    print(f"- Correlation between order value and rating: {corr_rating_spend:.3f}")

if set(["Shipping Type", "Review Rating"]).issubset(df.columns):
    print("\nAverage rating by shipping type:")
    print(df.groupby("Shipping Type")["Review Rating"]
            .mean()
            .round(2))
```

```python
# Age distribution
if "Age" in df.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df["Age"], bins=20, kde=True)
    plt.title("Customer age distribution")
    plt.xlabel("Age")
    plt.ylabel("Number of customers")
    plt.show()

# Spend by gender
if set(["Gender", "Purchase Amount (USD)"]).issubset(df.columns):
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x="Gender", y="Purchase Amount (USD)")
    plt.title("Purchase amount by gender")
    plt.show()

# Category popularity
if "Category" in df.columns:
    plt.figure(figsize=(8,4))
    top_cat = df["Category"].value_counts().head(10).index
    sns.countplot(data=df[df["Category"].isin(top_cat)], x="Category",
                  order=top_cat)
    plt.title("Top 10 categories by order count")
    plt.xticks(rotation=45, ha="right")
    plt.show()
```

```python
import numpy as np

# Work on a copy to keep original safe
data = df.copy()

# 1) Make column names simpler (optional but helpful)
data.columns = (
    data.columns
        .str.strip()
        .str.replace(" ", "_")
        .str.replace("(", "", regex=False)
        .str.replace(")", "", regex=False)
        .str.replace("/", "_")
)

print("Renamed columns:")
print(list(data.columns))
```

```python
# 2) Basic sanity checks on key numeric fields
numeric_cols = ["Age", "Purchase_Amount_USD", "Previous_Purchases", "Review_Rating"]
print("\nQuick numeric summary:")
print(data[numeric_cols].describe())
```

```python
# 3) Make a simple target for later: is_subscribed (1/0)
data["is_subscribed"] = (data["Subscription_Status"].str.strip().str.lower() == "yes").astype(int)

print("\nSubscription flag distribution:")
print(data["is_subscribed"].value_counts())
```

```python
# 4) Save this prepared version for later steps
prepared = data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Use the prepared copy
data = prepared.copy()

# 1) Choose features (exclude Customer_ID and is_subscribed for clustering)
feature_cols = [
    "Age",
    "Gender",
    "Item_Purchased",
    "Category",
    "Purchase_Amount_USD",
    "Location",
    "Size",
    "Color",
    "Season",
    "Review_Rating",
    "Shipping_Type",
    "Discount_Applied",
    "Promo_Code_Used",
    "Previous_Purchases",
    "Payment_Method",
    "Frequency_of_Purchases",
]

X = data[feature_cols]

```

```python
# 2) Identify numeric and categorical columns
numeric_features = ["Age", "Purchase_Amount_USD", "Previous_Purchases", "Review_Rating"]
categorical_features = [c for c in feature_cols if c not in numeric_features]

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)

```

```python
# 3) Build preprocessing pipeline
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)

```

```python
# 4) Fit the preprocessor and create the transformed matrix
X_prepared = preprocessor.fit_transform(X)

print("\nShape of prepared feature matrix:", X_prepared.shape)
```

```python
from sklearn.cluster import KMeans

# 1) Choose number of clusters (start with 4 for readable segments)
k = 4

kmeans = KMeans(
    n_clusters=k,
    random_state=42,
    n_init=10
)

```

```python
# 2) Fit on the prepared feature matrix
kmeans.fit(X_prepared)

```

```python
# 3) Assign cluster labels to each customer
data = prepared.copy()
data["segment"] = kmeans.labels_

print("Segment counts:")
print(data["segment"].value_counts().sort_index())

```

```python
# 4) Look at basic profile of each segment
segment_profile = data.groupby("segment").agg({
    "Age": "mean",
    "Purchase_Amount_USD": "mean",
    "Previous_Purchases": "mean",
    "Review_Rating": "mean",
    "is_subscribed": "mean"
}).round(2)

print("\nNumeric profile by segment:")
print(segment_profile)

```

```python
# 5) Check category and season preference per segment
print("\nTop category per segment:")
print(data.groupby("segment")["Category"].agg(lambda x: x.value_counts().index[0]))

print("\nTop season per segment:")
print(data.groupby("segment")["Season"].agg(lambda x: x.value_counts().index[0]))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Use the prepared data (with is_subscribed)
data = prepared.copy()

# Features (same as for clustering) + we can also include the cluster segment
feature_cols = [
    "Age",
    "Gender",
    "Item_Purchased",
    "Category",
    "Purchase_Amount_USD",
    "Location",
    "Size",
    "Color",
    "Season",
    "Review_Rating",
    "Shipping_Type",
    "Discount_Applied",
    "Promo_Code_Used",
    "Previous_Purchases",
    "Payment_Method",
    "Frequency_of_Purchases",
]

X = data[feature_cols]
y = data["is_subscribed"]

numeric_features = ["Age", "Purchase_Amount_USD", "Previous_Purchases", "Review_Rating"]
categorical_features = [c for c in feature_cols if c not in numeric_features]

```

```python
# Reuse the preprocessor definition
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocessor_clf = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)

```

```python
# Build full pipeline: preprocess -> logistic regression
clf = Pipeline(steps=[
    ("preprocess", preprocessor_clf),
    ("model", LogisticRegression(max_iter=2000))
])

```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

```

```python
# Fit model
clf.fit(X_train, y_train)

```

```python
# Evaluate
y_pred = clf.predict(X_test)

print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification report:")
print(classification_report(y_test, y_pred, digits=3))
```

```python
import numpy as np

# Get the trained logistic regression model and the one-hot encoder
log_reg = clf.named_steps["model"]
ohe = clf.named_steps["preprocess"].named_transformers_["cat"]

```

```python
# Build feature name list
num_features = numeric_features
cat_feature_names = ohe.get_feature_names_out(categorical_features)
all_feature_names = np.concatenate([num_features, cat_feature_names])

```

```python
# Get coefficients for class 1 (subscriber)
coefs = log_reg.coef_[0]

```

```python
# Top 15 positive and negative drivers
top_pos_idx = np.argsort(coefs)[-15:][::-1]
top_neg_idx = np.argsort(coefs)[:15]

```

```python
print("Top factors that INCREASE chance of subscription:")
for i in top_pos_idx:
    print(f"{all_feature_names[i]:40s}  coef={coefs[i]:.3f}")

print("\nTop factors that DECREASE chance of subscription:")
for i in top_neg_idx:
    print(f"{all_feature_names[i]:40s}  coef={coefs[i]:.3f}")
```
